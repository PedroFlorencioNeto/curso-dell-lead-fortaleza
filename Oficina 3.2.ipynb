```python
df = sns.load_dataset('mpg')
df.head()
```

<b>(Quest√£o 2) </b> Crie uma visualiza√ß√£o que mostre a rela√ß√£o entre a acelera√ß√£o (‚Äòacceleration‚Äô) e o consumo de combust√≠vel (‚Äòmpg‚Äô) dos ve√≠culos, em que seja poss√≠vel diferenciar o pa√≠s de origem (‚Äòorigin‚Äô). Sabendo disso, existe uma correla√ß√£o entre a acelera√ß√£o e o consumo? De que tipo? A correla√ß√£o √© a mesma para os diferentes pa√≠ses de origem? Voc√™ pode escolher a t√©cnica que julgar mais adequada!

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='acceleration',y='mpg',hue='origin',data=df)
plt.title('Gr√°fico de dispers√£o da acelera√ß√£o e consumo de ve√≠culos por pa√≠s de origem')
plt.show()
```

```python
sns.relplot(x='acceleration',y='mpg',col='origin',data=df)
plt.show()
```

```python
sns.heatmap(df.corr(),cmap='viridis',annot=True,linewidths=1)
plt.show()
```

üöó √â poss√≠vel notar que h√° uma baixa correla√ß√£o positiva entre consumo e acelera√ß√£o dos carros deste conjunto de dados, principalmente se observarmos apenas os ve√≠culos de origem europeia. A melhor visualiza√ß√£o fica por conta dos plots separados de cada pa√≠s de origem aliado √† matriz de correla√ß√£o. As correla√ß√µes entre cada um deles √© diferente. Vamos ver se os carros europeus tem realmente a menor correla√ß√£o?

```python
print('Correla√ß√µes\n')
print('Europa - {}'.format(df[df['origin']=='europe'].corr()['acceleration'].iloc[0]))
print('Jap√£o - {}'.format(df[df['origin']=='japan'].corr()['acceleration'].iloc[0]))
print('EUA - {}'.format(df[df['origin']=='usa'].corr()['acceleration'].iloc[0]))
```

Exatamente isso!

<b>(Quest√£o 3)</b> Crie uma visualiza√ß√£o que mostre a rela√ß√£o entre a acelera√ß√£o (‚Äòacceleration‚Äô) e o consumo de combust√≠vel (‚Äòmpg‚Äô) dos ve√≠culos, em que seja poss√≠vel visualizar o ano do modelo (model_year). Existe alguma rela√ß√£o entre as tr√™s vari√°veis?
