<b>(Quest√£o 2)</b> Crie uma visualiza√ß√£o que mostre a distribui√ß√£o do valor da passagem (‚Äúfare‚Äù), de acordo com a cidade na qual o passageiro embarcou ("embark_town"). Voc√™ pode escolher a t√©cnica que julgar mais adequada

* <b>Op√ß√£o 1: Gr√°fico de dispers√£o categ√≥rico</b>

```python
plt.figure(figsize=(7,7))
sns.set_style('whitegrid')
sns.stripplot(x='embark_town',y='fare',data=df,palette='pastel')
plt.title('Gr√°fico 1 - Valores pagos de passagem por cidade',fontdict={'fontsize':15},pad=25)
plt.xlabel('Cidade',fontdict={'fontsize':12},labelpad=20)
plt.ylabel('Valor da passagem',fontdict={'fontsize':12},labelpad=20)
plt.show()
```

* <b>Op√ß√£o 2: Gr√°fico de enxame (swarmplot)</b>

```python
plt.figure(figsize=(7,7))
sns.set_style('whitegrid')
sns.swarmplot(x='embark_town',y='fare',data=df,palette='pastel')
plt.title('Gr√°fico 2 - Valores pagos de passagem por cidade',fontdict={'fontsize':15},pad=25)
plt.xlabel('Cidade',fontdict={'fontsize':12},labelpad=20)
plt.ylabel('Valor da passagem',fontdict={'fontsize':12},labelpad=20)
plt.show()
```

* <b>Op√ß√£o 3: Gr√°fico de caixa (boxplot)</b>

```python
plt.figure(figsize=(7,7))
sns.set_style('whitegrid')
sns.boxplot(x='embark_town',y='fare',data=df,palette='pastel')
plt.title('Gr√°fico 3 - Valores pagos de passagem por cidade',fontdict={'fontsize':15},pad=25)
plt.xlabel('Cidade',fontdict={'fontsize':12},labelpad=20)
plt.ylabel('Valor da passagem',fontdict={'fontsize':12},labelpad=20)
plt.yticks(ticks=np.arange(0,550,25))
plt.show()
```

* <b>Op√ß√£o 4: Gr√°fico de caixa (boxplot) alternativo</b>

```python
plt.figure(figsize=(7,7))
sns.set_style('whitegrid')
sns.boxenplot(x='embark_town',y='fare',data=df,palette='pastel')
plt.title('Gr√°fico 4 - Valores pagos de passagem por cidade',fontdict={'fontsize':15},pad=25)
plt.xlabel('Cidade',fontdict={'fontsize':12},labelpad=20)
plt.ylabel('Valor da passagem',fontdict={'fontsize':12},labelpad=20)
plt.yticks(ticks=np.arange(0,550,25))
plt.show()
```

* <b>Op√ß√£o 5: Gr√°fico de violino (violinplot)</b>

```python
plt.figure(figsize=(7,7))
sns.set_style('whitegrid')
sns.violinplot(x='embark_town',y='fare',data=df,palette='pastel')
plt.title('Gr√°fico 5 - Valores pagos de passagem por cidade',fontdict={'fontsize':15},pad=25)
plt.xlabel('Cidade',fontdict={'fontsize':12},labelpad=20)
plt.ylabel('Valor da passagem',fontdict={'fontsize':12},labelpad=20)
plt.show()
```

<b> Conclus√µes: </b>
* Por conter muitos dados concentrados, o gr√°fico de enxame se mostrou melhor que o gr√°fico de dispers√£o categ√≥rico (stripplot). 
* O gr√°fico de violino por sua vez, deveria fornecer uma melhor visualiza√ß√£o que os anteriores, j√° que estamos lidando com muitos dados. Entretanto, isso n√£o ocorre para esse caso.
* Podemos observar atrav√©s do stripplot que Queenstown foi a cidade que menos pessoas embarcaram.
* A distribui√ß√£o de passagens mostra que h√° pouca dispers√£o no valor pago pelos passageiros, principalmente em Queenstown onde os valores dos quartis est√£o muito pr√≥ximos da mediana (e abaixo de 25 d√≥lares).
* Atrav√©s do boxplot, notamos que tivemos dezenas de passageiros s√£o considerados outliers, isto √©, pagaram passagem muito mais cara que foi pago pela maioria dos passageiros de suas respectivas cidades.
