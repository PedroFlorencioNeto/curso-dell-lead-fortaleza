```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt')
df.head()
```

```python
df.info()
```

<b>(Quest√£o 2) </b>Para o segundo e √∫ltimo passo, crie uma visualiza√ß√£o que mostre a s√©rie temporal das vendas ao longo dos meses. Voc√™ pode, por exemplo, se perguntar se existe alguma tend√™ncia ou alguma periodicidade, OK?

```python
plt.figure(figsize=(25,8))
sns.set_style('whitegrid')
plt.title('Vendas de Shampoos',fontdict={'fontsize':18})
sns.lineplot(x=df['Month'],y=df['Sales'],palette='inferno')
plt.xlabel('Meses',fontdict={'fontsize':14})
plt.ylabel('Vendas',fontdict={'fontsize':14})
plt.show()
```

<b> Observa√ß√µes sobre os dados: </b>

* As coletas de dados sobre as vendas de shampoos foram feitas ao longo de 3 anos.
* Houve uma tend√™ncia de crescimento nas vendas ao longo desses 3 anos.
* Entre julho a outubro do primeiro ano houve uma queda nas vendas, chegando muito pr√≥ximo do pior √≠ndice de vendas ao longo de todo o tempo analisado.
* O maior √≠ndice de vendas ocorreu no m√™s de setembro do terceiro ano.
